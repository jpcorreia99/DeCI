```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with_sleep = pd.read_csv("./data/with_sleep.csv",dtype="float")
```

```python
with_sleep
```

```python
without_sleep = pd.read_csv("./data/without_sleep.csv",dtype="float")
```

```python
without_sleep
```

# Total duration vs number of peers

```python
fig, ax = plt.subplots(figsize=(20,8),dpi=300)
ax.plot(with_sleep["number of peers"], with_sleep["total duration"])
ax.set_title("Computation duration evolution with increasing number of nodes (heavy computation)", fontsize=18)
ax.set_xlabel("Number of nodes", fontsize=18)
ax.set_ylabel("Duration (sec)", fontsize=18)
plt.savefig("./plots/dur_evo_from_1_heavy.png")
ax.plot()
```

```python
fig, ax = plt.subplots(figsize=(20,8),dpi=300)
ax.plot(with_sleep["number of peers"][1:], with_sleep["total duration"][1:])
ax.set_title("Computation duration evolution with increasing number of nodes starting at 10 peers (heavy computation)", fontsize=18)
ax.set_xlabel("Number of nodes", fontsize=18)
ax.set_ylabel("Duration (sec)", fontsize=18)
plt.savefig("./plots/dur_evo_from_10_heavy.png")
ax.plot()
```

```python
fig, ax = plt.subplots(figsize=(20,8),dpi=300)
ax.plot(without_sleep["number of peers"], without_sleep["total duration"])
ax.set_title("Computation duration evolution with increasing number of nodes (lite computation)", fontsize=18)
ax.set_xlabel("Number of nodes", fontsize=18)
ax.set_ylabel("Duration (sec)", fontsize=18)
plt.savefig("./plots/dur_evo_from_1_lite.png")
ax.plot()
```

```python
fig, ax = plt.subplots(figsize=(20,8),dpi=300)
ax.plot(without_sleep["number of peers"][1:], without_sleep["total duration"][1:])
ax.set_title("Computation duration evolution with increasing number of nodes starting at 10 peers (lite computation)", fontsize=18)
ax.set_xlabel("Number of nodes", fontsize=18)
ax.set_ylabel("Duration (sec)", fontsize=18)
plt.savefig("./plots/dur_evo_from_10_lite.png")
ax.plot()
```

# Pure computation vs paxos overhead

```python
fig, ax = plt.subplots(figsize=(20,8),dpi=300)
ax.plot(with_sleep["number of peers"], with_sleep["pure execution duration"])
ax.plot(with_sleep["number of peers"], with_sleep["paxos duration"])
ax.set_title("Pure computation time vs paxos overhead (heavy computation)", fontsize=18)
ax.set_xlabel("Number of nodes", fontsize=18)
ax.set_ylabel("Duration (sec)", fontsize=18)
ax.legend(["pure execution", "paxos overhead"], fontsize=18)
plt.savefig("./plots/pure_vs_paxos_from_1_heavy.png")
ax.plot()
```

```python
fig, ax = plt.subplots(figsize=(20,8),dpi=300)
ax.plot(with_sleep["number of peers"][1:], with_sleep["pure execution duration"][1:])
ax.plot(with_sleep["number of peers"][1:], with_sleep["paxos duration"][1:])
ax.set_title("Pure computation time vs paxos overhead starting at 10 peers (heavy computation)", fontsize=18)
ax.set_xlabel("Number of nodes", fontsize=18)
ax.set_ylabel("Duration (sec)", fontsize=18)
ax.legend(["pure execution", "paxos overhead"], fontsize=18)
plt.savefig("./plots/pure_vs_paxos_from_10_heavy.png")
ax.plot()
```

```python
fig, ax = plt.subplots(figsize=(20,8),dpi=300)
ax.plot(without_sleep["number of peers"], without_sleep["pure execution duration"])
ax.plot(without_sleep["number of peers"], without_sleep["paxos duration"])
ax.set_title("Pure computation time vs paxos overhead (lite computation)", fontsize=18)
ax.set_xlabel("Number of nodes", fontsize=18)
ax.set_ylabel("Duration (sec)", fontsize=18)
ax.legend(["pure execution", "paxos overhead"], fontsize=18)
plt.savefig("./plots/pure_vs_paxos_from_1_lite.png")
ax.plot()
```

```python
fig, ax = plt.subplots(figsize=(20,8),dpi=300)
ax.plot(without_sleep["number of peers"][1:], without_sleep["pure execution duration"][1:])
ax.plot(without_sleep["number of peers"][1:], without_sleep["paxos duration"][1:])
ax.set_title("Pure computation time vs paxos overhead starting at 10 peers (lite computation)", fontsize=18)
ax.set_xlabel("Number of nodes", fontsize=18)
ax.set_ylabel("Duration (sec)", fontsize=18)
ax.legend(["pure execution", "paxos overhead"], fontsize=18)
plt.savefig("./plots/pure_vs_paxos_from_10_lite.png")
ax.plot()
```

# Evolution of paxos overhead as a percentage of the total computation

```python
with_sleep["paxos percentage"] = with_sleep["paxos duration"] / with_sleep["total duration"]
```

```python
with_sleep["pure percentage"] = with_sleep["pure execution duration"] / with_sleep["total duration"]
```

```python
fig, ax = plt.subplots(figsize=(20,8),dpi=300)
with_sleep.plot(x="number of peers", y=["pure percentage","paxos percentage"], kind="bar", ax=ax)
ax.set_title("Pure computation time vs paxos overhead as percentages of total computation time (heavy computation)", fontsize=18)
ax.set_xlabel("Number of nodes", fontsize=18)
ax.set_ylabel("Duration (sec)", fontsize=18)
ax.legend(["pure execution", "paxos overhead"], fontsize=18)
plt.savefig("./plots/pure_vs_paxos_perce_heavy.png")
ax.plot()
```

```python
without_sleep["paxos percentage"] = without_sleep["paxos duration"] / without_sleep["total duration"]
```

```python
without_sleep["pure percentage"] = without_sleep["pure execution duration"] / without_sleep["total duration"]
```

```python
fig, ax = plt.subplots(figsize=(20,8),dpi=300)
without_sleep.plot(x="number of peers", y=["pure percentage","paxos percentage"], kind="bar", ax=ax)
ax.set_title("Pure computation time vs paxos overhead as percentages of total computation time (lite computation)", fontsize=18)
ax.set_xlabel("Number of nodes", fontsize=18)
ax.set_ylabel("Duration (sec)", fontsize=18)
ax.legend(["pure execution", "paxos overhead"], fontsize=18)
plt.savefig("./plots/pure_vs_paxos_perce_lite.png")
ax.plot()
```
